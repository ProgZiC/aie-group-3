```python
# # HW10-11: Computer Vision in PyTorch (оптимизировано для слабого железа)

# ## Часть A: классификация (CIFAR100)
# ## Часть B: детекция (Pascal VOC)


# ### 1. Импорты, seed, устройство

import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
import os
from tqdm import tqdm

# Фиксация seed для воспроизводимости
seed = 42
torch.manual_seed(seed)
np.random.seed(seed)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

# Устройство: используем CPU, если нет GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Создание папок для артефактов
os.makedirs("artifacts/figures", exist_ok=True)

# Для ускорения на CPU уменьшаем число рабочих процессов
num_workers = 0 if device.type == 'cpu' else 2


# ### 2. Часть A: классификация CIFAR100 (оптимизировано)


# #### 2.1. Загрузка данных

# Базовые трансформации для CNN (32x32)
transform_base_cnn = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

# Аугментации для CNN (упрощённые, без ColorJitter для скорости)
transform_aug_cnn = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

# Трансформации для ResNet (96x96 вместо 224, чтобы было быстрее)
transform_base_resnet = transforms.Compose([
    transforms.Resize(96),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

transform_aug_resnet = transforms.Compose([
    transforms.RandomResizedCrop(96),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

# Загрузка CIFAR100
print("Загрузка CIFAR100...")
full_train_cnn = torchvision.datasets.CIFAR100(root='./data', train=True, download=True, transform=transform_base_cnn)
test_dataset_cnn = torchvision.datasets.CIFAR100(root='./data', train=False, download=True, transform=transform_base_cnn)

full_train_resnet = torchvision.datasets.CIFAR100(root='./data', train=True, download=True, transform=transform_base_resnet)
test_dataset_resnet = torchvision.datasets.CIFAR100(root='./data', train=False, download=True, transform=transform_base_resnet)

# Разделение train на train (80%) и val (20%)
train_size = int(0.8 * len(full_train_cnn))
val_size = len(full_train_cnn) - train_size
train_dataset_cnn, val_dataset_cnn = random_split(full_train_cnn, [train_size, val_size],
                                                  generator=torch.Generator().manual_seed(seed))

# Для аугментированного обучения CNN
full_train_aug_cnn = torchvision.datasets.CIFAR100(root='./data', train=True, download=True, transform=transform_aug_cnn)
train_dataset_aug_cnn, _ = random_split(full_train_aug_cnn, [train_size, val_size],
                                        generator=torch.Generator().manual_seed(seed))

# Разделение для ResNet (те же индексы, чтобы split был одинаковым)
indices = list(range(len(full_train_resnet)))
train_indices, val_indices = indices[:train_size], indices[train_size:]
train_dataset_resnet = torch.utils.data.Subset(full_train_resnet, train_indices)
val_dataset_resnet = torch.utils.data.Subset(full_train_resnet, val_indices)

# Для аугментированного обучения ResNet
full_train_aug_resnet = torchvision.datasets.CIFAR100(root='./data', train=True, download=True, transform=transform_aug_resnet)
train_dataset_aug_resnet = torch.utils.data.Subset(full_train_aug_resnet, train_indices)

# DataLoaders (уменьшенный batch_size для слабого железа)
batch_size = 32  # было 64, уменьшили для экономии памяти
train_loader_cnn = DataLoader(train_dataset_cnn, batch_size=batch_size, shuffle=True, num_workers=num_workers)
val_loader_cnn = DataLoader(val_dataset_cnn, batch_size=batch_size, shuffle=False, num_workers=num_workers)
test_loader_cnn = DataLoader(test_dataset_cnn, batch_size=batch_size, shuffle=False, num_workers=num_workers)
train_loader_aug_cnn = DataLoader(train_dataset_aug_cnn, batch_size=batch_size, shuffle=True, num_workers=num_workers)

train_loader_resnet = DataLoader(train_dataset_resnet, batch_size=batch_size, shuffle=True, num_workers=num_workers)
val_loader_resnet = DataLoader(val_dataset_resnet, batch_size=batch_size, shuffle=False, num_workers=num_workers)
test_loader_resnet = DataLoader(test_dataset_resnet, batch_size=batch_size, shuffle=False, num_workers=num_workers)
train_loader_aug_resnet = DataLoader(train_dataset_aug_resnet, batch_size=batch_size, shuffle=True, num_workers=num_workers)


# #### 2.2. Sanity‑check


images, labels = next(iter(train_loader_cnn))
print(f"CNN train batch shape: {images.shape}, labels: {labels.shape}")

# Визуализация нескольких изображений
def imshow(img, title=None):
    img = img.permute(1,2,0).numpy()
    img = img * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
    img = np.clip(img, 0, 1)
    plt.imshow(img)
    if title:
        plt.title(title)
    plt.axis('off')

fig, axes = plt.subplots(1, 5, figsize=(12,3))
for i in range(5):
    img = images[i]
    img = img.permute(1,2,0).numpy()
    img = img * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
    img = np.clip(img, 0, 1)
    axes[i].imshow(img)
    axes[i].set_title(f"Label: {labels[i].item()}")
    axes[i].axis('off')
plt.tight_layout()
plt.show()


# #### 2.3. Простая CNN


class SimpleCNN(nn.Module):
    def __init__(self, num_classes=100):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 32, 3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, 3, padding=1)
        self.conv3 = nn.Conv2d(64, 128, 3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        # После трёх пулингов: 32 -> 16 -> 8 -> 4
        self.fc1 = nn.Linear(128 * 4 * 4, 512)
        self.fc2 = nn.Linear(512, num_classes)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        x = self.pool(self.relu(self.conv1(x)))
        x = self.pool(self.relu(self.conv2(x)))
        x = self.pool(self.relu(self.conv3(x)))
        x = x.view(x.size(0), -1)
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x


# #### 2.4. Функции обучения и оценки

def train_one_epoch(model, loader, optimizer, criterion, device):
    model.train()
    total_loss = 0
    correct = 0
    total = 0
    for inputs, targets in tqdm(loader, desc="Training"):
        inputs, targets = inputs.to(device), targets.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()
        total_loss += loss.item() * inputs.size(0)
        _, predicted = outputs.max(1)
        total += targets.size(0)
        correct += predicted.eq(targets).sum().item()
    return total_loss / total, correct / total

def evaluate(model, loader, criterion, device):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, targets in tqdm(loader, desc="Evaluating"):
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            total_loss += loss.item() * inputs.size(0)
            _, predicted = outputs.max(1)
            total += targets.size(0)
            correct += predicted.eq(targets).sum().item()
    return total_loss / total, correct / total

def run_experiment(model, train_loader, val_loader, epochs=15, lr=0.001, save_name=None):
    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    best_val_acc = 0
    history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
    for epoch in range(epochs):
        train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, criterion, device)
        val_loss, val_acc = evaluate(model, val_loader, criterion, device)
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        print(f"Epoch {epoch+1}: train_loss={train_loss:.4f}, train_acc={train_acc:.4f}, val_loss={val_loss:.4f}, val_acc={val_acc:.4f}")
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            if save_name:
                torch.save(model.state_dict(), f"{save_name}.pt")
    return history, best_val_acc


# #### 2.5. Эксперименты C1–C4


epochs = 15  # уменьшили с 20 для скорости
lr = 0.001

# C1: SimpleCNN без аугментаций
print("=== C1: SimpleCNN без аугментаций ===")
model_c1 = SimpleCNN()
hist_c1, best_val_c1 = run_experiment(model_c1, train_loader_cnn, val_loader_cnn, epochs, lr, save_name="best_c1")
print(f"C1 best val acc: {best_val_c1:.4f}")

# C2: SimpleCNN с аугментациями
print("=== C2: SimpleCNN с аугментациями ===")
model_c2 = SimpleCNN()
hist_c2, best_val_c2 = run_experiment(model_c2, train_loader_aug_cnn, val_loader_cnn, epochs, lr, save_name="best_c2")
print(f"C2 best val acc: {best_val_c2:.4f}")

# C3: ResNet18 с замороженным backbone
print("=== C3: ResNet18 с замороженным backbone ===")
model_c3 = torchvision.models.resnet18(pretrained=True)
for param in model_c3.parameters():
    param.requires_grad = False
num_ftrs = model_c3.fc.in_features
model_c3.fc = nn.Linear(num_ftrs, 100)
hist_c3, best_val_c3 = run_experiment(model_c3, train_loader_resnet, val_loader_resnet, epochs, lr, save_name="best_c3")
print(f"C3 best val acc: {best_val_c3:.4f}")

# C4: ResNet18 с дообучением последних слоёв (layer4 + fc)
print("=== C4: ResNet18 с дообучением layer4 + fc ===")
model_c4 = torchvision.models.resnet18(pretrained=True)
for param in model_c4.parameters():
    param.requires_grad = False
for param in model_c4.layer4.parameters():
    param.requires_grad = True
model_c4.fc = nn.Linear(num_ftrs, 100)
model_c4.fc.requires_grad = True
hist_c4, best_val_c4 = run_experiment(model_c4, train_loader_resnet, val_loader_resnet, epochs, lr, save_name="best_c4")
print(f"C4 best val acc: {best_val_c4:.4f}")


# #### 2.6. Выбор лучшей модели по валидации и тестирование


best_val_accs = {
    "C1": best_val_c1,
    "C2": best_val_c2,
    "C3": best_val_c3,
    "C4": best_val_c4
}
best_model_id = max(best_val_accs, key=best_val_accs.get)
print(f"Best model: {best_model_id} with val_acc = {best_val_accs[best_model_id]:.4f}")

# Загружаем лучшие веса и тестируем
if best_model_id in ["C1", "C2"]:
    model_best = SimpleCNN()
    model_best.load_state_dict(torch.load(f"best_{best_model_id.lower()}.pt"))
    test_loader = test_loader_cnn
else:
    model_best = torchvision.models.resnet18(pretrained=False)
    num_ftrs = model_best.fc.in_features
    model_best.fc = nn.Linear(num_ftrs, 100)
    if best_model_id == "C3":
        for param in model_best.parameters():
            param.requires_grad = False
    else:  # C4
        for param in model_best.parameters():
            param.requires_grad = False
        for param in model_best.layer4.parameters():
            param.requires_grad = True
        model_best.fc.requires_grad = True
    model_best.load_state_dict(torch.load(f"best_{best_model_id.lower()}.pt"))
    test_loader = test_loader_resnet

model_best = model_best.to(device)
test_loss, test_acc = evaluate(model_best, test_loader, nn.CrossEntropyLoss(), device)
print(f"Test accuracy of best model: {test_acc:.4f}")


# #### 2.7. Сохранение артефактов классификации


# Создаём DataFrame для runs.csv
results = []
for exp in ["C1","C2","C3","C4"]:
    val_acc = best_val_accs[exp]
    results.append({
        "experiment_id": exp,
        "task": "classification",
        "dataset": "CIFAR100",
        "seed": seed,
        "model_summary": "SimpleCNN" if exp in ["C1","C2"] else "ResNet18",
        "optimizer": "Adam",
        "lr": lr,
        "epochs_trained": epochs,
        "best_val_accuracy": val_acc,
        "test_accuracy": test_acc if exp == best_model_id else None,
        "precision": None,
        "recall": None,
        "mean_iou": None,
        "notes": ""
    })
df_results = pd.DataFrame(results)

# Сохраняем best_classifier.pt и конфиг
torch.save(model_best.state_dict(), "artifacts/best_classifier.pt")
config = {
    "dataset": "CIFAR100",
    "architecture": "ResNet18" if best_model_id in ["C3","C4"] else "SimpleCNN",
    "transforms": {
        "train": "transform_aug_cnn" if best_model_id == "C2" else ("transform_aug_resnet" if best_model_id == "C4" else "transform_base_cnn"),
        "val": "transform_base_cnn" if best_model_id in ["C1","C2"] else "transform_base_resnet",
        "test": "transform_base_cnn" if best_model_id in ["C1","C2"] else "transform_base_resnet"
    },
    "seed": seed,
    "hyperparameters": {
        "lr": lr,
        "epochs": epochs,
        "batch_size": batch_size,
        "optimizer": "Adam"
    }
}
with open("artifacts/best_classifier_config.json", "w") as f:
    json.dump(config, f, indent=2)

# График кривых лучшего прогона
if best_model_id == "C1":
    best_hist = hist_c1
elif best_model_id == "C2":
    best_hist = hist_c2
elif best_model_id == "C3":
    best_hist = hist_c3
else:
    best_hist = hist_c4

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(best_hist['train_loss'], label='train loss')
plt.plot(best_hist['val_loss'], label='val loss')
plt.legend()
plt.title('Loss curves')
plt.subplot(1,2,2)
plt.plot(best_hist['train_acc'], label='train acc')
plt.plot(best_hist['val_acc'], label='val acc')
plt.legend()
plt.title('Accuracy curves')
plt.savefig("artifacts/figures/classification_curves_best.png")
plt.show()

# Сравнение C1-C4
plt.figure()
plt.bar(list(best_val_accs.keys()), list(best_val_accs.values()), color=['blue','green','orange','red'])
plt.ylabel('Best validation accuracy')
plt.title('Comparison of experiments C1-C4')
plt.savefig("artifacts/figures/classification_compare.png")
plt.show()

# Визуализация аугментаций (на примере CNN аугментаций)
def show_augmentations(dataset, num_images=5):
    fig, axes = plt.subplots(1, num_images, figsize=(15,3))
    for i in range(num_images):
        img, _ = dataset[i]
        img = img.permute(1,2,0).numpy()
        img = img * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
        img = np.clip(img, 0, 1)
        axes[i].imshow(img)
        axes[i].axis('off')
    plt.savefig("artifacts/figures/augmentations_preview.png")
    plt.show()

show_augmentations(train_dataset_aug_cnn)


# ### 3. Часть B: детекция на Pascal VOC (оптимизировано)


# #### 3.1. Загрузка датасета и модели


# Загружаем предобученную Faster R-CNN
model_det = torchvision.models.detection.fasterrcnn_resnet50_fpn(pretrained=True)
model_det.eval()
model_det.to(device)

# Загружаем датасет VOC Detection (2012) – только для оценки, не для обучения
train_det = torchvision.datasets.VOCDetection(root='./data', year='2012', image_set='train', download=True)
val_det = torchvision.datasets.VOCDetection(root='./data', year='2012', image_set='val', download=True)

# Для sanity-check
sample_img, sample_target = val_det[0]
print(f"Sample image size: {sample_img.size}")
print(f"Sample target keys: {sample_target.keys()}")


# #### 3.2. Вспомогательные функции


def compute_iou(boxA, boxB):
    # box: [xmin, ymin, xmax, ymax]
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
    interArea = max(0, xB - xA) * max(0, yB - yA)
    boxAArea = (boxA[2]-boxA[0]) * (boxA[3]-boxA[1])
    boxBArea = (boxB[2]-boxB[0]) * (boxB[3]-boxB[1])
    iou = interArea / (boxAArea + boxBArea - interArea)
    return iou

def get_gt_boxes(target):
    """Извлекает все bounding boxes из аннотации VOC."""
    boxes = []
    if 'annotation' in target:
        objects = target['annotation']['object']
        if not isinstance(objects, list):
            objects = [objects]
        for obj in objects:
            bbox = obj['bndbox']
            boxes.append([int(bbox['xmin']), int(bbox['ymin']), int(bbox['xmax']), int(bbox['ymax'])])
    return np.array(boxes) if boxes else np.empty((0,4))

def compute_metrics(model, dataset, score_threshold=0.5, iou_threshold=0.5, device='cpu', max_samples=30):
    tp = 0
    fp = 0
    fn = 0
    total_iou = 0
    matched = 0
    indices = range(len(dataset)) if max_samples is None else range(min(max_samples, len(dataset)))
    for idx in tqdm(indices, desc=f"Computing metrics (th={score_threshold})"):
        img, target = dataset[idx]
        img_tensor = transforms.ToTensor()(img).unsqueeze(0).to(device)
        with torch.no_grad():
            pred = model(img_tensor)[0]
        # Предсказанные боксы с уверенностью выше порога
        keep = pred['scores'] > score_threshold
        pred_boxes = pred['boxes'][keep].cpu().numpy()
        # GT боксы
        gt_boxes = get_gt_boxes(target)

        # Простой greedy match
        matched_gt = set()
        for pb in pred_boxes:
            best_iou = 0
            best_idx = -1
            for j, gb in enumerate(gt_boxes):
                iou = compute_iou(pb, gb)
                if iou > best_iou:
                    best_iou = iou
                    best_idx = j
            if best_iou >= iou_threshold and best_idx not in matched_gt:
                tp += 1
                matched_gt.add(best_idx)
                total_iou += best_iou
                matched += 1
            else:
                fp += 1
        fn += len(gt_boxes) - len(matched_gt)

    precision = tp / (tp + fp) if (tp+fp) > 0 else 0
    recall = tp / (tp + fn) if (tp+fn) > 0 else 0
    mean_iou = total_iou / matched if matched > 0 else 0
    return precision, recall, mean_iou


# #### 3.3. Оценка для V1 и V2 (ограниченное количество примеров)


# Используем небольшую часть validation для ускорения
max_samples = 30  # можно увеличить, если есть ресурсы

# V1: threshold = 0.3
prec1, rec1, miou1 = compute_metrics(model_det, val_det, score_threshold=0.3, device=device, max_samples=max_samples)
print(f"V1 (th=0.3): Precision={prec1:.4f}, Recall={rec1:.4f}, mIoU={miou1:.4f}")

# V2: threshold = 0.7
prec2, rec2, miou2 = compute_metrics(model_det, val_det, score_threshold=0.7, device=device, max_samples=max_samples)
print(f"V2 (th=0.7): Precision={prec2:.4f}, Recall={rec2:.4f}, mIoU={miou2:.4f}")


# #### 3.4. Визуализация предсказаний


def visualize_detections(model, dataset, idx, thresholds=[0.3, 0.7], device='cpu'):
    img, target = dataset[idx]
    img_tensor = transforms.ToTensor()(img).unsqueeze(0).to(device)
    with torch.no_grad():
        pred = model(img_tensor)[0]

    fig, axes = plt.subplots(1, len(thresholds)+1, figsize=(15,5))
    # GT
    axes[0].imshow(img)
    axes[0].set_title('Ground Truth')
    gt_boxes = get_gt_boxes(target)
    for box in gt_boxes:
        rect = plt.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1],
                             fill=False, edgecolor='green', linewidth=2)
        axes[0].add_patch(rect)
    axes[0].axis('off')

    for i, th in enumerate(thresholds):
        keep = pred['scores'] > th
        boxes = pred['boxes'][keep].cpu().numpy()
        scores = pred['scores'][keep].cpu().numpy()
        axes[i+1].imshow(img)
        axes[i+1].set_title(f'Predictions (th={th})')
        for box, score in zip(boxes, scores):
            rect = plt.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1],
                                 fill=False, edgecolor='red', linewidth=2)
            axes[i+1].add_patch(rect)
            axes[i+1].text(box[0], box[1]-5, f'{score:.2f}', color='red', fontsize=8)
        axes[i+1].axis('off')
    plt.tight_layout()
    return fig

# Визуализируем несколько примеров (ограничимся 3)
example_indices = [0, 10, 20]
for idx in example_indices:
    fig = visualize_detections(model_det, val_det, idx, thresholds=[0.3, 0.7], device=device)
    plt.show()

# Сохраняем примеры в файл
fig, axes = plt.subplots(len(example_indices), 3, figsize=(15, 4*len(example_indices)))
for row, idx in enumerate(example_indices):
    img, target = val_det[idx]
    img_tensor = transforms.ToTensor()(img).unsqueeze(0).to(device)
    with torch.no_grad():
        pred = model_det(img_tensor)[0]

    # GT
    axes[row,0].imshow(img)
    axes[row,0].set_title('Ground Truth')
    gt_boxes = get_gt_boxes(target)
    for box in gt_boxes:
        rect = plt.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1],
                             fill=False, edgecolor='green', linewidth=2)
        axes[row,0].add_patch(rect)
    axes[row,0].axis('off')

    # Pred th=0.3
    keep = pred['scores'] > 0.3
    boxes = pred['boxes'][keep].cpu().numpy()
    scores = pred['scores'][keep].cpu().numpy()
    axes[row,1].imshow(img)
    axes[row,1].set_title('th=0.3')
    for box, score in zip(boxes, scores):
        rect = plt.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1],
                             fill=False, edgecolor='red', linewidth=2)
        axes[row,1].add_patch(rect)
        axes[row,1].text(box[0], box[1]-5, f'{score:.2f}', color='red', fontsize=8)
    axes[row,1].axis('off')

    # Pred th=0.7
    keep = pred['scores'] > 0.7
    boxes = pred['boxes'][keep].cpu().numpy()
    scores = pred['scores'][keep].cpu().numpy()
    axes[row,2].imshow(img)
    axes[row,2].set_title('th=0.7')
    for box, score in zip(boxes, scores):
        rect = plt.Rectangle((box[0], box[1]), box[2]-box[0], box[3]-box[1],
                             fill=False, edgecolor='red', linewidth=2)
        axes[row,2].add_patch(rect)
        axes[row,2].text(box[0], box[1]-5, f'{score:.2f}', color='red', fontsize=8)
    axes[row,2].axis('off')

plt.tight_layout()
plt.savefig("artifacts/figures/detection_examples.png")
plt.show()


# #### 3.5. Сохранение метрик детекции в runs.csv


det_results = [
    {
        "experiment_id": "V1",
        "task": "detection",
        "dataset": "VOC2012",
        "seed": seed,
        "model_summary": "FasterRCNN_ResNet50_FPN",
        "optimizer": None,
        "lr": None,
        "epochs_trained": None,
        "best_val_accuracy": None,
        "test_accuracy": None,
        "precision": prec1,
        "recall": rec1,
        "mean_iou": miou1,
        "notes": f"score_threshold=0.3, IoU_threshold=0.5, samples={max_samples}"
    },
    {
        "experiment_id": "V2",
        "task": "detection",
        "dataset": "VOC2012",
        "seed": seed,
        "model_summary": "FasterRCNN_ResNet50_FPN",
        "optimizer": None,
        "lr": None,
        "epochs_trained": None,
        "best_val_accuracy": None,
        "test_accuracy": None,
        "precision": prec2,
        "recall": rec2,
        "mean_iou": miou2,
        "notes": f"score_threshold=0.7, IoU_threshold=0.5, samples={max_samples}"
    }
]

# Добавляем к ранее сохранённым результатам классификации
df_results = pd.concat([df_results, pd.DataFrame(det_results)], ignore_index=True)
df_results.to_csv("artifacts/runs.csv", index=False)


# #### 3.6. Визуализация сравнения метрик


labels = ['Precision', 'Recall', 'mean IoU']
v1_vals = [prec1, rec1, miou1]
v2_vals = [prec2, rec2, miou2]
x = np.arange(len(labels))
width = 0.35
fig, ax = plt.subplots()
ax.bar(x - width/2, v1_vals, width, label='V1 (th=0.3)')
ax.bar(x + width/2, v2_vals, width, label='V2 (th=0.7)')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.title('Detection metrics comparison')
plt.savefig("artifacts/figures/detection_metrics.png")
plt.show()
```
